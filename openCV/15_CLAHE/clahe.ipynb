```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('giraffe-kenya.png')

if image is None:
    print('The image is not loaded sucessfully')
else:
    print('The image is loaded sucessfully')
```

```python
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
equalized_image = cv2.equalizeHist(image_gray)

clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(4,4))
clahe_equalized_image = clahe.apply(image_gray)


plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
plt.title('Original image')
plt.imshow(image_gray, cmap='grey')

plt.subplot(1,3,2)
plt.title('equalized histogram image')
plt.imshow(equalized_image,cmap='grey')

plt.subplot(1,3,3)
plt.title('clahe iamge')
plt.imshow(clahe_equalized_image,cmap='grey')
plt.show()
```

```python
image = cv2.imread('giraffe-kenya.png')

yuv_image = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
yuv_image[:,:,0] = clahe.apply(yuv_image[:,:,0])

yuv_image_BGR = cv2.cvtColor(yuv_image,cv2.COLOR_YUV2BGR)
yuv_image_RGB = cv2.cvtColor(yuv_image_BGR, cv2.COLOR_BGR2RGB)

image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.title('original image')
plt.imshow(image_rgb)

plt.subplot(1,2,2)
plt.title('clahe image')
plt.imshow(yuv_image_RGB)

plt.show()
```
