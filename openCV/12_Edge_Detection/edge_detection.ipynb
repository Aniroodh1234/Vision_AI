```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('giraffe-Kenya.png',cv2.IMREAD_GRAYSCALE)
```

#### Performing edge dectection

#### In sobel edge detection, the x axis is determined by a separate filter as of y. Then we combine both the images and find the magnitude to get the edge

```python
### Applying sobel edge detection

sobel_edge_x = cv2.Sobel(image,cv2.CV_64F,1,0,ksize=3)  ## here 1 mewans edge is determined along the x direction
sobel_edge_y = cv2.Sobel(image,cv2.CV_64F,0,1,ksize=3)  ## here 1 mewans edge is determined along the y direction

sobel_combined = cv2.magnitude(sobel_edge_x,sobel_edge_y)

### Applying laplacian edge detection

laplacian_edge = cv2.Laplacian(image,cv2.CV_64F,ksize=3)
laplacian_abs = cv2.convertScaleAbs(laplacian_edge)   ## in laplacian edge detection we will get both positive and negative values, 
                                                ## in order to remove all negative values we find the absolute

## Applying canny
canny_edge = cv2.Canny(image,180,200)   ## in canny edge the edge is measured on a threshold scale, 
                                        ## 180 is the lower threshold and 200 is the higher threshold
```

```python
titles = ['Original', 'Sobel', 'Laplacian', 'Laplacian_abs', 'Canny']
image_list = [image, sobel_combined, laplacian_edge, laplacian_abs,  canny_edge]

plt.figure(figsize=(20,16))
for i in range(len(titles)):
    plt.subplot(3,2, i+1)
    plt.imshow(image_list[i], cmap='grey')
    plt.title(titles[i])
    plt.axis('off')
plt.show()

```
