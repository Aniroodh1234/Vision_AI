```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
images = cv2.imread('filter.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(images,cmap='gray')
```

```python
x = -10

horizontal_filter = np.array([[1,1,1],
                                [x,x,x],
                                [-1,-1,-1]])

vertical_filter = np.array([[1,x,-1],
                            [1,x,-1],
                            [1,x,-1]])

diagonal_filter = np.array([[x,-1,-1],
                            [1,x,-1],
                            [1,1,x]])

horizontal_filter_image = cv2.filter2D(images,-1,horizontal_filter)
vertical_filter_image = cv2.filter2D(images,-1,vertical_filter)
diagonal_filter_image = cv2.filter2D(images,-1,diagonal_filter)

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title('original image')
plt.imshow(images,cmap='gray')
plt.axis('off')

plt.subplot(2,2,2)
plt.title('Horizontal Filter')
plt.imshow(horizontal_filter_image,cmap='gray')
plt.axis('off')

plt.subplot(2,2,3)
plt.title('Vertical Filter')
plt.imshow(vertical_filter_image,cmap='gray')
plt.axis('off')

plt.subplot(2,2,4)
plt.title('Diagonal Filter')
plt.imshow(diagonal_filter_image,cmap='gray')
plt.axis('off')

plt.show()

```
