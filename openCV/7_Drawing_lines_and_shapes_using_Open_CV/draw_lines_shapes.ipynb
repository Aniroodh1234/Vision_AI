```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Create a empty canvas
height,width = 500,700
black = (0,0,0)
canvas = np.full((height,width,3),black,dtype=np.uint8)

plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

# Draw to a line

```python
cv2.line(canvas,(50,50),(650,450),(255,255,255),4)  # 4 is the thickness of the line

plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

# Draw a rectangle

```python
cv2.rectangle(canvas,(100,100),(300,300),(0,255,255),-1)

plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

# Draw a circle

```python
cv2.circle(canvas, (500, 150), 100, (0, 0, 255), 4)

plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))

```

# Draw an Ellipse

```python
cv2.ellipse(canvas, (350, 350), (100, 50), 45, 0, 360, (0, 255, 0), 3)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

# Draw a polygon

```python
# Xn, Yn, where n>2
## here we need to create a 3D numpy array which mention all the sides

point = np.array([[[150,400],[300,400],[225,300]]],np.int32)
new_reshape_points = point.reshape(-1,1,2)

print(new_reshape_points)

cv2.polylines(canvas, [new_reshape_points], False, (255, 0, 255), 3)  ## false indicates that the first point and last point of the polydon is not connected
                                                                        ## to connect it we neeed to use true

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```
