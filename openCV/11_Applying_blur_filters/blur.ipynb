```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

#### blurring is always done on grayscale images, so we need to take input as grayscale and want output as grayscale

```python
image = cv2.imread('giraffe-Kenya.png')
image_grey = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
kernel_s = 10
average_blur = cv2.blur(image_grey,(kernel_s,kernel_s))   ## always we use kernel size in yuple for average blurring
plt.imshow(average_blur, cmap='grey')
```

```python
kernel_s = 31  # in gaussian blur the kernel size should be odd and greater than 1

gaussian_blur = cv2.GaussianBlur(image_grey, (kernel_s, kernel_s), 0,0)  # the first 0 is standard deviation in X direction and the second 0 is standard deviation in Y direction, more the sigma value more the blurring effect

plt.imshow(gaussian_blur, cmap='grey')
```

```python
median_blur = cv2.medianBlur(image_grey, 11)  # in median blur we should single kernel size and it should be odd and greater than 1

plt.imshow(median_blur, cmap='grey')
```

```python
plt.figure(figsize=(14, 12))
plt.subplot(2,2,1)
plt.title('original image')
plt.imshow(image_grey, cmap='grey')
plt.axis('off')

plt.subplot(2,2,2)
plt.title('average blur image')
plt.imshow(average_blur, cmap='grey')
plt.axis('off')

plt.subplot(2,2,3)
plt.title('gaussian blur image')
plt.imshow(gaussian_blur, cmap='grey')
plt.axis('off')

plt.subplot(2,2,4)
plt.title('median blur image')
plt.imshow(median_blur, cmap='grey')
plt.axis('off')
```

#### median blurring is mainly used for smoothening the image and removing the noise form images
#### gaussian blur and average blur is used to do the blurring of photos
