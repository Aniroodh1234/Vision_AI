```python


image = cv2.imread('jirrafe.jpeg')
```

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(image_rgb)
```

```python
## here we are targetting the white portion of the girrafe. Hence we have created the upper bound and lower bound accoording to it
## always the upper and lower bound are taken as numpy array

lower_bound = np.array([0,0,200])
uppper_bound = np.array([30,80,255])
```

```python


mask = cv2.inRange(image_hsv,lower_bound,uppper_bound)
```

```python
plt.imshow(mask)
```

```python
## here we are creating the color which is used in place of white color of the girrafe 


black_hsv = np.uint8([[[120,255,0]]])
black_bgr = cv2.cvtColor(black_hsv,cv2.COLOR_HSV2BGR)[0][0]
```

```python
## always we will perform, experiments on the cpopy of original image

image_result = image.copy()
image_result[mask>0] = black_bgr
```

```python
image_result_rgb = cv2.cvtColor(image_result,cv2.COLOR_BGR2RGB)
```

```python
# PLOT

fig,axs = plt.subplots(1,3,figsize=(12,6))

axs[0].imshow(image_rgb)
axs[0].set_title("Original Image")
axs[0].axis('off')

axs[1].imshow(mask)
axs[1].set_title("mask image")
axs[1].axis('off')

axs[2].imshow(image_result_rgb)
axs[2].set_title("output Image")
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
