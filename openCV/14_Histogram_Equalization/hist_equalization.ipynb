```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('giraffe-Kenya.png')

if image is None:
    print('Image is not loaded sucessfully')
else:
    print("image is loaded sucessfully")
```

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_grayscale = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
hist,bins = np.histogram(image_grayscale.flatten(),256,[0,256])  ## before equalizing histogram we will first flattern

equalized_image = cv2.equalizeHist(image_grayscale)

equalised_hist,bins = np.histogram(equalized_image.flatten(),256,[0,256])  ## after equalizing we will flattern the image

plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
plt.title('original image')
plt.imshow(image_grayscale,cmap='grey')

plt.subplot(2,2,2)
plt.title('original image histogram')
plt.plot(hist,color='blue')

plt.subplot(2,2,3)
plt.title('equalized image')
plt.imshow(equalized_image,cmap='grey')

plt.subplot(2,2,4)
plt.title('equalized image histogram')
plt.plot(equalised_hist,color='red')

plt.tight_layout()
plt.show()
```

#### histogram equalization can only be done using a single channel, more preferably a grayscale image

```python
image_yuv = cv2.cvtColor(image,cv2.COLOR_BGR2YUV)

image_yuv[:,:,0] = cv2.equalizeHist(image_yuv[:,:,0])
equalized_color_image = cv2.cvtColor(image_yuv,cv2.COLOR_YUV2BGR)
equalized_rgb_image = cv2.cvtColor(equalized_color_image,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,8))

plt.subplot(1,2,1)
plt.title("original image")
plt.imshow(image_rgb)

plt.subplot(1,2,2)
plt.title("equalized RGB image")
plt.imshow(equalized_rgb_image)

plt.tight_layout()
plt.show()
```
