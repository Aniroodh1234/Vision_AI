```python
import cv2
import matplotlib.pyplot as plt
```

```python


image = cv2.imread('jirrafe.jpeg')
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

# Flip

```python
# 1 --> horizontally
# 0 --> Vertically
# -1 --> Both axes


flip_horizontal = cv2.flip(image,1)
flip_vertical = cv2.flip(image,0)
flip_both = cv2.flip(image,-1)

title = ['image_rgb','flip_horizontal','flip_vertical','flip_both']
image = [image_rgb,
        cv2.cvtColor(flip_horizontal,cv2.COLOR_BGR2RGB),
        cv2.cvtColor(flip_vertical,cv2.COLOR_BGR2RGB),
        cv2.cvtColor(flip_both,cv2.COLOR_BGR2RGB)
        ]
plt.figure(figsize=(6,4))
for i in range(4):
  plt.subplot(2,2,i+1)
  plt.imshow(image[i])
  plt.title(title[i])
  plt.axis('off')

plt.tight_layout()
plt.show()
```

# Rotate

```python
rotate_90_clockwise = cv2.rotate(image_rgb,cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterclockwise = cv2.rotate(image_rgb,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv2.rotate(image_rgb,cv2.ROTATE_180)

title = ['original image','rotate 90 - clockwise','rotate 90 - counterclockwise','rotate 180']
images = [image_rgb,rotate_90_clockwise,rotate_90_counterclockwise,rotate_180]

plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(title[i])
    plt.axis('on')
plt.tight_layout()
plt.show()

```

# Cropping and image

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
print(type(image_rgb))
print(image_rgb.shape)
```

```python
x_start, y_start = 160,150
x_end, y_end = 300,280
cropped_image = image_rgb[y_start:y_end, x_start:x_end]

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title('original image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cropped_image)
plt.title('cropped image')
plt.axis('off')
```
